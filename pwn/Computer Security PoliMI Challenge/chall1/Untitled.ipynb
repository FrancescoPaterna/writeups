```python
def buffer_searcher(pointer, pill): 
    a = 10*pointer
    b = 10*(pointer+1)
    
    for i in range(a,b):
        pill = pill + str('%'+ str(i)+'$x ')
    
    return pill
```

```python
pill = ''
```

```python
pill = buffer_searcher(3, pill)
```

```python
print(pill)
```

```python
f = open("pill.txt", "w")
f.write(pill)
f.close()
```

```python
import os
```

```python
os.system("./mission1 < pill.txt")
```

```python
for i in range (5000):
    pill = buffer_searcher(1, pill)
    f = open("pill.txt", "w")
    f.write(pill)
    f.close()
    os.system("./mission1 < pill.txt")
```
